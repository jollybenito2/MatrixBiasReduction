```python
def get_data_from_yahoo(tickers):
    start = dt.datetime(2019, 1, 1)
    end = dt.datetime(2022, 4, 30)
    for ticker in tickers:
        # just in case your connection breaks, we'd like to save our progress!
        if not os.path.exists('BMV_Days/{}.csv'.format(ticker.rstrip("\n"))):
            #df = pdr.DataReader(ticker.rstrip("\n"), start, 
            #           end, data_source='yahoo')['Adj Close']
            df = yf.download(ticker.rstrip("\n"), start, end, "1m")      
            df.reset_index(inplace=True)
            df.set_index("Date", inplace=True)
            # Check for empty df and do not save them.
            if df.empty:
                print('DataFrame is empty!')
            else: 
                df.to_csv('BMV_Days/{}.csv'.format(ticker.rstrip("\n")))
        else:
            print('Already have {}'.format(ticker.rstrip("\n")))

get_data_from_yahoo(list(Stocks1['Acciones']))
```

```python
def get_data_from_yahoo(tickers):
    dates = [dt.datetime(2022, 4, 1), dt.datetime(2022, 4, 7),
             dt.datetime(2022, 4, 14), dt.datetime(2022, 4, 21)]
    for ticker in tickers:
        # just in case your connection breaks, we'd like to save our progress!
        if not os.path.exists('BMV_Minutes_1/{}.csv'.format(ticker.rstrip("\n"))):
            #df = pdr.DataReader(ticker.rstrip("\n"), start, 
            #           end, data_source='yahoo')['Adj Close']
            df1 = yf.download(ticker.rstrip("\n"), dates[0], dates[1], interval = "1m")      
            df2 = yf.download(ticker.rstrip("\n"), dates[1], dates[2], interval = "1m")
            df3 = yf.download(ticker.rstrip("\n"), dates[2], dates[3], interval = "1m")      
            #df4 = yf.download(ticker.rstrip("\n"), dates[1], dates[2], interval = "1m")
            df = pd.concat([df1,df2,df3])     
            
            df.reset_index(inplace=True)
            df.set_index("Datetime", inplace=True)
            # Check for empty df and do not save them.
            if df.empty:
                print('DataFrame is empty!')
            else: 
                df.to_csv('BMV_Minutes_1/{}.csv'.format(ticker.rstrip("\n")))
        else:
            print('Already have {}'.format(ticker.rstrip("\n")))

get_data_from_yahoo(list(Stocks1['Acciones']))
```
